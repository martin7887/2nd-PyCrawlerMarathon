### 1. 取出知乎問題發問時間

```python
# 1. 取出知乎問題發問時間

import requests

headers = {
  'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.122 Safari/537.36'
}
r = requests.get('https://www.zhihu.com/api/v4/questions/55493026/answers',headers=headers)
response = r.text
response
```

```python
import json
import datetime

response_json = json.loads(response)
for key in response_json['data'][0]['question'].keys():
    print(key,":",response_json['data'][0]['question'][key])
    
timeStamp = response_json['data'][0]['question']['created']
dateArray = datetime.datetime.utcfromtimestamp(timeStamp)
otherStyleTime = dateArray.strftime('%Y-%m-%d %H:%M:%S')

print('知乎問題發問時間:',otherStyleTime)  
```

### 2. 取出第一筆與最後一筆回答的時間

```python
# 2. 取出第一筆與最後一筆回答的時間
first_page = 'https://www.zhihu.com/api/v4/questions/55493026/answers'
last_page = 'https://www.zhihu.com/api/v4/questions/55493026/answers?limit=5&offset=555'
headers = {
  'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.122 Safari/537.36'
}
r = requests.get(last_page,headers=headers)
response = r.text
```

```python
import requests
import time
import random
import json

headers = {
  'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.122 Safari/537.36'
}

time_list=[]
def find_create_time(response_json):
    for i in range(len(response_json['data'])):
        timeStamp = response_json['data'][i]['created_time']
        dateArray = datetime.datetime.utcfromtimestamp(timeStamp)
        otherStyleTime = dateArray.strftime('%Y-%m-%d %H:%M:%S')
        time_list.append(otherStyleTime)
        
url = 'https://www.zhihu.com/api/v4/questions/55493026/answers'
response = requests.get(url,headers=headers).text
response_json = json.loads(response)
total = response_json['paging']['totals']
find_create_time(response_json)
        
for offset in range(5,total,5):
    url = f'https://www.zhihu.com/api/v4/questions/55493026/answers?limit=5&offset={offset}'
    response = requests.get(url,headers=headers).text
    response_json = json.loads(response)
    find_create_time(response_json)
    time.sleep(random.random())
time_list
```

```python
date = [x.split(' ')[0] for x in time_list]
time = [x.split(' ')[1] for x in time_list]
import pandas as pd
df = (pd.DataFrame({'Date':date,
                   'Time':time
                  })).sort_values('Date').reset_index()
df.iloc[[0,-1],[1,2]]
```
