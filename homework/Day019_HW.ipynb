```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
webdriver_path = '/Users/Martin/Downloads/python_github/第二屆Python網路爬蟲馬拉松/課程資料/chromedriver'
# browser = webdriver.Chrome(executable_path='chromedriver')
browser = webdriver.Chrome(executable_path=webdriver_path)
browser.get("http://taqm.epa.gov.tw/taqm/tw/MonthlyAverage.aspx")
```

```python
#選取 測站：高雄小港，年度：2019
selectSite = Select(browser.find_element_by_id("ctl05_ddlSite"))
selectSite.select_by_value('58')
selectYear = Select(browser.find_element_by_id("ctl05_ddlYear"))
selectYear.select_by_value('2019')

#按下查詢按鈕
browser.find_element_by_id('ctl05_btnQuery').click()
```

![](https://imgur.com/LwGGT8c.jpg)

```python
# 取得資料（等瀏覽器資料出現後才執行）
html_source = browser.page_source
html_source
```

```python
soup = BeautifulSoup(html_source, 'lxml')
table = soup.find('table', class_='TABLE_G')
table
```

```python
Kaohsiung_air_data=dict()
Monitor_items = []
units = []
dates = []
values = []

for i in range(0,12*10,12):
    Monitor_items.append(table.find_all('tr',style = "color:Black;")[i].find_all('td')[0].text)
    units.append(table.find_all('tr',style = "color:Black;")[i].find_all('td')[1].text)
    dates.append(table.find_all('tr',style = "color:Black;")[i].find_all('td')[2].text)
    values.append(table.find_all('tr',style = "color:Black;")[i].find_all('td')[3].text)
    for j in range(1,12):
        dates.append(table.find_all('tr',style = "color:Black;")[j].find_all('td')[0].text)
        values.append(table.find_all('tr',style = "color:Black;")[j].find_all('td')[1].text)

for i in range(10):
    Monitor_item = Monitor_items[i]
    Kaohsiung_air_data[Monitor_item]=dict(zip(dates[i*12:(i+1)*12:1],values[i*12:(i+1)*12:1]))
Kaohsiung_air_data
    
#     
```
