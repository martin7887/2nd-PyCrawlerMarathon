```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
Kaohsiung_air_data=dict()
Monitor_items = []
units = []
dates = []
values = []

for i in range(0,12*10,12):
    Monitor_items.append(table.find_all('tr',style = "color:Black;")[i].find_all('td')[0].text)
    units.append(table.find_all('tr',style = "color:Black;")[i].find_all('td')[1].text)
    dates.append(table.find_all('tr',style = "color:Black;")[i].find_all('td')[2].text)
    values.append(table.find_all('tr',style = "color:Black;")[i].find_all('td')[3].text)
    for j in range(1,12):
        dates.append(table.find_all('tr',style = "color:Black;")[j].find_all('td')[0].text)
        values.append(table.find_all('tr',style = "color:Black;")[j].find_all('td')[1].text)

for i in range(10):
    Monitor_item = Monitor_items[i]
    Kaohsiung_air_data[Monitor_item]=dict(zip(dates[i*12:(i+1)*12:1],values[i*12:(i+1)*12:1]))
Kaohsiung_air_data
    
#     
```
